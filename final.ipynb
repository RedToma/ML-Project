```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dd = pd.read_csv('used-car.csv')
plt.figure(figsize=(12, 6))
sns.countplot(dd["year"])
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x = dd['year'])
```

```python
ax2 = sns.countplot(x = dd['transmission'])
```

```python
sns.barplot(x = dd['transmission'], y = dd['price'], ci = False)
```

```python
plt.scatter(x = dd['year'], y = dd['price'])
```

```python
sns.distplot(dd['mileage'], color="y")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(x = dd['mileage'], y = dd['price'], c = 'green')
plt.show()
```

```python
plt.figure(figsize=(15,10),facecolor='w') 
sns.scatterplot(dd["mileage"], dd["price"], hue = dd["year"])
```
